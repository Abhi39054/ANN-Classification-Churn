```python
import tensorflow as tf
import pickle
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
```

```python
model = load_model("model.h5")

# Load encoder decoder

with open("label_encoder_gender.pkl", "rb") as file:
    label_enocder_gender = pickle.load(file)

with open("ohe_encoder_geography.pkl", "rb") as file:
    ohe_encoder_geography = pickle.load(file)

with open("scaler.pkl", "rb") as file:
    scaler = pickle.load(file)
```

```python
input_data = {
    "CreditScore":600,
    "Geography":"France",
    "Gender":"Male",
    "Age":40,
    "Tenure":3,
    "Balance":60000,
    "NumOfProducts":2,
    "HasCrCard":1,
    "IsActiveMember":1,
    "EstimatedSalary":50000
}
```

```python
input_data_test = {
    "CreditScore":[600],
    "Geography":["France"],
    "Gender":["Male"],
    "Age":[40],
    "Tenure":[3],
    "Balance":[60000],
    "NumOfProducts":[2],
    "HasCrCard":[1],
    "IsActiveMember":[1],
    "EstimatedSalary":[50000]
}
data = pd.DataFrame(input_data_test)

# Apply label encoder to Gender
data["Gender"] = label_enocder_gender.transform(data["Gender"])

# Apply OneHotEncoder on Geography

geo_df = pd.DataFrame(ohe_encoder_geography.transform([[input_data["Geography"]]]), columns=ohe_encoder_geography.get_feature_names_out())
data = pd.concat([data.drop("Geography", axis=1), geo_df], axis=1)


# Apply Standard Scaler
X = scaler.transform(data)

X = pd.DataFrame(X, columns=data.columns)


# apply prediction
probabilities = model.predict(X)

# GEt the actual output class
predicted_classes = (probabilities > 0.5).astype("int32")

print("Predicted Class = ", predicted_classes)
```
